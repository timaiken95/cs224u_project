```python
import pandas as pd
import re
import numpy as np
import vsm
```

```python
vectors = np.load("GloVe_codeswitch_5k.npy")
words = np.load('5k_vocab_dict.npy').item()
embedding_dict = {}
for k,v in words.items():
    embedding_dict[k] = vectors[v]
```

```python
df = pd.DataFrame.from_dict(embedding_dict, orient="index")
print(df)
```

```python
vsm.tsne_viz(df, output_filename="tsne-viz-2.png", figsize=(40, 50))
```
